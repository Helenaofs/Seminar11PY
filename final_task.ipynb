```python
import sympy
import numpy as np
from sympy.plotting import plot
from sympy import symbols
```

```python
x = symbols('x')
plot(-12*x**4*sympy.sin(sympy.cos(x)) - 18*x**3 + 5*x**2 + 10*x - 30, (x, -100, 100))
```

```python
sympy.solveset(sympy.Eq(-12*x**4*sympy.sin(sympy.cos(x)) - 18*x**3 + 5*x**2 + 10*x - 30, 0), x)
```

```python
x = np.arange(-10, 10, 1)
f = -12*x**4 * (np.sin(np.cos(x))) - 18*x**3 + 5*x**2 + 10*x - 30

max_list = []
min_list = []

for i in x:
    res = -12*i**4 * (np.sin(np.cos(i))) - 18*i**3 + 5*i**2 + 10*i - 30
    if res > 0:
        max_list.append(res)
    else:
        min_list.append(res)

print(f'f > 0: {max_list[-1]} --> {max_list[0]}')
print(max_list)
print(f'f < 0:{min_list[0]} --> {min_list[-1]}')
print(min_list)
```
